the code below represents how to load kaggle datasets to colab without manually uploading it.<br>
by using this method it took shorter time to load kaggle datasets.

before any steps you need to upload kaggle.json<br>
if you don't have kaggle.json you need to go [kaggle settings](https://www.kaggle.com/settings) of your own profile and then scroll down to the API, then click on "Expire Token" and then select "Create New Token".<br>
once you do this a new kaggle.json file will be downloaded to your computer.

```python
# install kaggle library
! pip install kaggle

#creating a directory named kaggle
! mkdir ~/.kaggle

# copying kaggle.json to new directory
! cp kaggle.json ~/.kaggle/

! chmod 600 ~/.kaggle/kaggle.json

# downloading dataset
# note that to download kaggle competition you need to mention ! kaggle competitions download -c 'name-of-competition'
## !"just copy API command which can found on kaggle dataset webpage" FOR EXAMPLE
! kaggle competitions download -c 'spaceship-titanic'

#creating a new directory in order to unzip dataset overthere
## !mkdir "Your Dataset name" FOR EXAMPLE
!mkdir spaceship-titanic

## !unzip "folder name you previously created.zip" -d//content/"Your Dataset name" FOR EXAMPLE
!unzip 'spaceship-titanic.zip' -d /content/spaceship-titanic
```
